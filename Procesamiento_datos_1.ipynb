```python
days = ['16-09-23.csv','16-09-24.csv','16-09-25.csv','16-09-26.csv','16-09-27.csv','16-09-28.csv',
        '16-09-29.csv','16-09-30.csv','16-10-01.csv','16-10-02.csv','16-10-03.csv','16-10-04.csv',
        '16-10-05.csv','16-10-06.csv','16-10-07.csv','16-10-08.csv','16-10-09.csv','16-10-10.csv',
        '16-10-11.csv','16-10-12.csv']

devices = pd.read_csv('original/devices_list.csv',sep=";")

for day in days:
    dataset = pd.read_csv('original/'+day)
    data_src = pd.merge(dataset, devices, on=['eth.src'], how='inner')
    data_src['eth.src'] = data_src['Device_ID']
    data = data_src[['Packet ID','TIME','Size','eth.src','eth.dst_x','IP.src','IP.dst','IP.proto','port.src','port.dst']]
    data.rename(columns={'eth.dst_x':'eth.dst'}, inplace=True)
    data.rename(columns={'eth.src':'device_src'}, inplace=True)
    
    data_dst = pd.merge(data, devices, on=['eth.dst'], how='inner')
    data_dst['eth.dst'] = data_src['Device_ID']
    data_dst.rename(columns={'eth.dst':'device_dest'}, inplace=True)
    data_dst[['Packet ID','TIME','Size','device_src','device_dest','IP.src','IP.dst','IP.proto','port.src','port.dst']].to_csv('dataset/crude/f-'+day)
    
```
